```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.bikewale.com/best-bikes-in-india/")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
soup
```

```python
all=soup.find_all("div",{"class":"o-f o-aw o-o o-lT o-mf o-ne o-nn  o-mO o-ha o-hA o-iD o-kY o-nl"})   #class="z_8FgQ o-hl"# o-jK o-ei o-fu #class="o-f o-aw o-o o-lT o-mf o-ne o-nn  o-mO o-ha o-hA o-iD o-kY o-nl

all
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("div",{"class":"KzDlHZ"})[0].text)
    print(item.find_all("div",{"class":"Nx9bqj _4b5DiR"})[0].text)
    print(item.find_all("ul",{"class":"G4BRas"})[0].text)
```

```python
l=[]
for item in all:   #to display all price value in the website
    d={}
    d["phone_name"]=item.find_all("div",{"class":"KzDlHZ"})[0].text
   
    d["phone_price"]=item.find_all("div",{"class":"Nx9bqj _4b5DiR"})[0].text.replace("₹","Rs.")
    d["phone_description"]=item.find_all("ul",{"class":"G4BRas"})[0].text
    l.append(d)
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
```
